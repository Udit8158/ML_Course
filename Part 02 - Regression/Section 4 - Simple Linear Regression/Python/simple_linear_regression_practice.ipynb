# Practice Simple Linear Regression

## Importing libaries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

## Import dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
print(dataset.head(2))
```

## Seperate features and labels

```python
X = dataset.iloc[:,: -1].values  # data from predict
y = dataset.iloc[:,-1:].values  # need to predice
print(X.shape)
print(y.shape)
```

## Seperate testing and training sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2,random_state=0)
```

## Train model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting on X_test and X_train

```python
y_predict_test = regressor.predict(X_test)
y_predict_train = regressor.predict(X_train)
```

## Visiualizing training set results

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, y_predict_train, color = "blue")
plt.title('Years of Experience vs Salary (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualizing test set results

```python
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_train, y_predict_train, color = "blue")
plt.title('Years of Experience vs Salary (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
regressor.predict([[1]]) # prediction of salary for 1 Y of exp employee.
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```
